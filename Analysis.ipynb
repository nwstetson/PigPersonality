# Analyzing Pig Personality

```python
# load library
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# import datasets
svy = pd.read_csv("CaregiverSurveyMOCKCleaner.csv")
obs = pd.read_csv("")
```

Overview of Personality Scores

```python
svy.describe()
```

```python
# Data
x = (svy["Pig"]) # x is the individual pig
x = np.sort(x, axis = 0) # sort array for line graph

plt.style.use('seaborn-poster') # customize the style
plt.figure(figsize=(6, 6))

# separate line for each personality trait ðŸ˜³
plt.subplot(111)
plt.plot(x, (svy["SC1"]), label = 'Fear of People');
plt.plot(x, (svy["SC2"]), label = 'Nonsocial Fear')
plt.plot(x, (svy["SC3"]), label = 'Fear of Pigs')
plt.plot(x, (svy["SC4"]), label = 'Fear of Handling')
plt.plot(x, (svy["SC5"]), label = 'General Aggression')
plt.plot(x, (svy["SC6"]), label = 'Situational Aggression')
plt.plot(x, (svy["SC7"]), label = 'Excitability')
plt.plot(x, (svy["SC8"]), label = 'Playfulness')
plt.plot(x, (svy["SC9"]), label = 'Active Engagement')
plt.plot(x, (svy["SC10"]), label = 'Companionability')
plt.plot(x, (svy["SC11"]), label = 'Aggression Towards Pigs')
plt.plot(x, (svy["SC12"]), label = 'Dominance Over Other Pigs')


plt.xlabel('Pig', fontsize = 18); # customize font size
plt.ylabel('Personality Scores', fontsize = 18); # customize font size
plt.title('Differences in Personality Scores by Pig', fontsize = 20,); # customize font size


plt.legend(bbox_to_anchor =(1.0, 0.37), fontsize = 16); # add legend


# change to scatter plot
# show tick for each pig
# change tick number to pig name?
```
